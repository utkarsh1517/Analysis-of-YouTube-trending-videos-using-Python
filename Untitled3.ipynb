```python
df.drop(['video_id','thumbnail_link'],axis=1,inplace=True)
```

```python
df['description_length']=df['description'].apply(len)
df['title_length']=df['title'].apply(len)
df['like_ratio']=df['likes']/df['views']
df['comment_ratio']=df['comment_count']/df['views']
df['dislike_ratio']=df['dislikes']/df['views']
```

```python
df['publish_day']=df['publish_time'].dt.day
df['publish_month']=df['publish_time'].dt.month
df['publish_year']=df['publish_time'].dt.year
df['publish_hours']=df['publish_time'].dt.hour
```

```python
df.drop(['publish_time'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['channel_title_encoded']= le.fit_transform(df['channel_title'])
```

```python
df.drop('channel_title',axis=1,inplace=True)
```

```python
df = pd.get_dummies(df,columns=['category_id'],prefix='cat',drop_first=True)
```
